Question b: Count the number of punctuation characters and remove punctuations.

```python
import string
count_punct = sum(1 for char in str if char in string.punctuation)
print("Number of punctuation characters:", count_punct) #for each punctuation mark, 1 gets added in the vriable count_punct
```

```python
# Removing punctuation
str_cleaned = str.translate(str.maketrans('', '', string.punctuation))

print("The entered paragraph without punctuation:")
print(str_cleaned)
```

```python
str_no_punct = word_tokenize(str_cleaned)
str_no_punct
```

Question c: Plot the distribution of stopwords used in the text using barplot.

```python
from nltk.corpus import stopwords
from collections import Counter #It creates a dictionary-like object where the elements of the collection are stored as keys, and their respective counts are stored as values.
import matplotlib.pyplot as plt

stopwords_list = stopwords.words('english')
stopwords_count = Counter(word for word in str_no_punct if word.lower() in stopwords_list)
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
plt.bar(stopwords_count.keys(), stopwords_count.values(), color='green')
plt.xlabel('Stopwords')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels
plt.tight_layout()
plt.show()
```

Question d: Remove the stopwords from the text.

```python
filtered_tokens = [word for word in str_no_punct if word.lower() not in stopwords_list]
```

```python
filtered_tokens
```

Question e: Plot the distribution of each POS Tag using a barplot.

```python
from nltk import pos_tag
pos_tags = pos_tag(filtered_tokens)
pos_tags_count = Counter(tag for word, tag in pos_tags)
plt.bar(pos_tags_count.keys(), pos_tags_count.values())
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Question f: Determine the number of unique lemma available in the text.

```python
from nltk.stem import WordNetLemmatizer
l = WordNetLemmatizer()
unique_lemmas = set(l.lemmatize(word) for word in filtered_tokens)
print("Number of unique lemmas:", len(unique_lemmas))
```

Question g: Plot the frequency distribution of words in the text. Display only the top 10
(most occurring) tokens in the chart.

```python
word_frequency = Counter(filtered_tokens)
top_10_words = word_frequency.most_common(10)
print("Top 10 most occurring words:", top_10_words)
```

```python
# Extract words and their frequencies
words = [word for word, _ in top_10_words]
frequencies = [freq for _, freq in top_10_words]

# Plot the top 10 most occurring words
plt.figure(figsize=(10, 6))
plt.barh(words, frequencies, color='skyblue')
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Words', fontsize=12)
plt.title('Top 10 Most Occurring Words', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to display highest frequency at the top
plt.tight_layout()
plt.show()
```

Question i: Find all dates and convert them to the DD-MM-YYYY format

```python
import re


dates = re.findall(r'\b(\d{1,2})-(?:January|February|March|April|May|June|July|August|September|October|November|December)-(\d{4})\b', str)
converted_dates = [f'{day}-03-{year}' for day, year in dates]  # Assuming the month is always March

print("Converted dates:")
print(converted_dates)
```

Question k: Determine whether the text contains any phone numbers in it. Ensure the phone
numbers are valid if any. If an invalid phone number is found, remove it from
the text.

```python
phone_numbers = re.findall(r'\+\d{11}\b', str)
phone_numbers
```

```python
# Remove phone numbers from the original text
for phone_number in phone_numbers:
    str = re.sub(re.escape(phone_number), '', str)

print("Text with phone numbers removed:\n")
print(str)
```
